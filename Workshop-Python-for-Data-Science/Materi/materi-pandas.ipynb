#### Perbedaan `.loc[]` vs `.iloc[]` di Pandas

| **Fitur**           | **.loc[]**                                | **.iloc[]**                        |
|---------------------|--------------------------------------|----------------------------------|
| **Berdasarkan**     | Label indeks (bisa angka atau string) | Posisi indeks (hanya angka)     |
| **Batas atas slicing** | **Inklusif** (batas atas termasuk) | **Eksklusif** (batas atas tidak termasuk) |
| **Bisa pakai string?** | ✅ Bisa | ❌ Tidak bisa |
| **Bisa pakai angka?** | ✅ Bisa (jika label indeks angka) | ✅ Bisa (karena posisi selalu angka) |

#### Catatan Penting: Hindari *Chained Indexing*

Saat kamu ingin **mengubah nilai** di DataFrame berdasarkan kondisi, jangan lakukan pola seperti ini:

```python
df[df["payment"].isna()]["payment"] = "unknown"   # ❌ berpotensi tidak bekerja / warning
```

Kenapa?
- `df[mask]` bisa menghasilkan **copy**, bukan view
- assignment-nya jadi tidak jelas nempel ke DataFrame yang mana

✅ Pola aman:
```python
mask = df["payment"].isna()
df.loc[mask, "payment"] = "unknown"
```

````python
# Contoh assignment aman dengan .loc (pola yang sering dipakai)
demo = data.copy()

# Isi missing payment dengan label "unknown"
mask = demo["payment"].isna()
demo.loc[mask, "payment"] = "unknown"

demo[["payment", "pickup_zone"]].head(10)   
````

- `passengers` >= 3

## **Pandas untuk Feature Engineering (Bonus DS)**

Di Data Science, kita sering membuat fitur baru dari:
- **Text** (kolom string) → pakai accessor `.str`
- **Datetime** (kolom waktu) → pakai accessor `.dt`

Di bawah contoh kecil yang sangat umum.


### Text Ops dengan `.str` (contoh: cari keyword di zona)

Yang sering dipakai di Data Science:
- `.str.contains()` untuk filtering
- `.str.lower()` / `.str.upper()` untuk normalisasi
- `.str.split()` untuk pecah teks

Referensi resmi:
- Text data & `.str`: https://pandas.pydata.org/docs/user_guide/text.html

````python
# Contoh .str untuk teks: bikin fitur "is_airport_pickup"
data_str = data.copy()

zone = data_str["pickup_zone"].fillna("")
data_str["is_airport_pickup"] = zone.str.contains("Airport", case=False, na=False)

data_str[["pickup_zone", "is_airport_pickup"]].head(10)
````

### Datetime Ops dengan `.dt` (langsung pakai kolom `pickup`)

Dataset kita sudah punya `pickup` & `dropoff`, jadi kita bisa langsung bikin fitur waktu:
- jam (hour)
- hari (day name)
- bulan (month)

Referensi resmi:
- Time series & `.dt`: https://pandas.pydata.org/docs/user_guide/timeseries.html

````python
data_time = data.copy()

data_time["pickup_hour"] = data_time["pickup"].dt.hour
data_time["pickup_day_name"] = data_time["pickup"].dt.day_name()
data_time["pickup_month"] = data_time["pickup"].dt.month

data_time[["pickup", "pickup_hour", "pickup_day_name", "pickup_month"]].head(10)
````

#### `NaN` vs `pd.NA` (Konsep Missing Value di Pandas)

Di Pandas, missing value bisa muncul sebagai:
- `np.nan` (umumnya untuk float)
- `pd.NA` (missing value yang “universal” untuk banyak dtype modern seperti `Int64`, `string`, `boolean`)

Catatan penting:
- `pd.NA` **tidak punya nilai boolean True/False** (ambiguous), jadi jangan dipakai langsung di `if`/kondisi tanpa `isna()`.

Referensi resmi:
- Missing data: https://pandas.pydata.org/docs/user_guide/missing_data.html


````python
s = pd.Series([1, pd.NA, 3], dtype="Int64")
print("Series dengan pd.NA:")
print(s)

print("\nCek missing:")
print(s.isna())

# Contoh operasi perbandingan tetap bisa (hasilnya juga bisa mengandung <NA>)
print("\nOperasi s > 1:")
print(s > 1)
````

> Catatan: biasanya kamu **memilih salah satu** strategi (drop *atau* fill). Contoh di bawah dipisah memakai variabel baru agar `data` asli tetap aman.

### **Menangani Data Duplikat**  

Terkadang, dataset mengandung **baris yang terduplikasi**, yang bisa mempengaruhi hasil analisis.  
Kita bisa menggunakan `.duplicated()` untuk mendeteksi baris yang muncul lebih dari satu kali.  

````python
# Dataset ini hampir tidak punya duplicate rows.
# Jadi untuk latihan, kita buat duplikat kecil secara sengaja:
df_dupes = pd.concat([data, data.sample(10, random_state=42)], ignore_index=True)
df_dupes.duplicated().sum()
````

````python
df_dupes_clean = df_dupes.drop_duplicates()
print("Duplikat sebelum:", df_dupes.duplicated().sum())
print("Duplikat setelah :", df_dupes_clean.duplicated().sum())
````

### **Menghapus Baris & Kolom dalam DataFrame**  

Kita bisa menghapus baris atau kolom yang tidak diperlukan dalam DataFrame menggunakan `.drop()`.  

````python
copy_data = data.copy()
copy_data.head()
````

#### **🔹 Parameter Utama:**  
1. **labels** → Label baris atau kolom yang ingin dihapus.  
   - Bisa berupa **satu label** atau **daftar label**.  
   - Jika menggunakan **tuple**, akan dianggap sebagai satu label, bukan daftar.  

2. **axis** → Menentukan apakah yang akan dihapus adalah **baris atau kolom**.  
   - `0` atau `'index'` → Menghapus **baris** berdasarkan label.  
   - `1` atau `'columns'` → Menghapus **kolom** berdasarkan label.  

3. **index** → Alternatif untuk menentukan baris yang dihapus.  
   - `data.drop(index=labels)` sama dengan `data.drop(labels, axis=0)`.  

4. **columns** → Alternatif untuk menentukan kolom yang dihapus.  
   - `data.drop(columns=labels)` sama dengan `data.drop(labels, axis=1)`.  


````python
# Hapus beberapa kolom
copy_data.drop(columns=["color", "tolls"], inplace=True)

# Hapus beberapa baris berdasarkan index
copy_data.drop(index=[1, 3], inplace=True)

copy_data.head()
````

### **GroupBy & Aggregation**  

#### **📌 Apa Itu `GroupBy`?**  
`GroupBy` adalah operasi yang memungkinkan kita untuk **mengelompokkan data** berdasarkan satu atau lebih kolom.  
Ini sangat berguna untuk melihat pola dalam dataset dengan cara mengorganisir data dalam kelompok tertentu.  

#### **📌 Apa Itu Aggregation?**  
Aggregation adalah proses **meringkas data** dengan menghitung statistik tertentu dari setiap kelompok hasil `GroupBy`.  
Contoh fungsi agregasi yang umum digunakan:  
- **`sum()`** → Menjumlahkan nilai dalam kelompok.  
- **`mean()`** → Menghitung rata-rata dalam kelompok.  
- **`count()`** → Menghitung jumlah elemen dalam kelompok.  
- **`max()` / `min()`** → Mencari nilai maksimum atau minimum dalam kelompok.  

#### 📌 Cara mikir `groupby`: Split → Apply → Combine

- **Split**: data dibagi berdasarkan kategori (misal `Produk`, `Cabang`, `Metode_Bayar`)
- **Apply**: lakukan fungsi (mean/sum/count/median/...)
- **Combine**: gabungkan hasilnya jadi tabel ringkas

````python
# Contoh transform: bikin kolom baru "TotalRevenuePerProduk"
df_sales = sales_df.copy()
df_sales["TotalRevenuePerProduk"] = df_sales.groupby("Produk")["Revenue_Rp"].transform("sum")
df_sales
````

### **Menggunakan `.apply()` dalam Pandas**  

Metode **`.apply()`** memungkinkan kita untuk menerapkan **fungsi kustom** (_user-defined function_) pada setiap baris atau kolom dalam DataFrame.  
Ini berguna untuk melakukan operasi yang lebih kompleks daripada metode bawaan Pandas seperti `.sum()` atau `.mean()`.

````python
def revenue_bucket(rev):
    if rev >= 350000:
        return "High"
    elif rev >= 250000:
        return "Medium"
    else:
        return "Low"

sales_df["Revenue_Bucket"] = sales_df["Revenue_Rp"].apply(revenue_bucket)
sales_df
````

````python
# 1) Row-wise apply (contoh) — biasanya LEBIH LAMBAT dari operasi vektorisasi
def calc_revenue_row(row):
    return row["Qty"] * row["Harga_Rp"]

sales_apply = sales_df.copy()
sales_apply["Revenue_apply"] = sales_apply.apply(calc_revenue_row, axis=1)

# Bandingkan dengan kolom Revenue_Rp yang kita buat vektorisasi
sales_apply[["Qty", "Harga_Rp", "Revenue_Rp", "Revenue_apply"]].head()
````

## Bonus DS: Menggabungkan Data (`concat` vs `merge`)

Kamu bakal sering ketemu kasus data kepisah di banyak tabel/file.

- `pd.concat([...])` → **numpuk** (row-wise/column-wise), cocok untuk *append dataset* yang strukturnya sama
- `pd.merge(left, right, on=...)` → **join** berdasarkan key, mirip SQL JOIN

Referensi resmi:
- Merging/Joining: https://pandas.pydata.org/docs/user_guide/merging.html


````python
# Contoh concat (numpuk row)
df_a = pd.DataFrame({"id":[1,2], "score":[10,20]})
df_b = pd.DataFrame({"id":[3], "score":[30]})

concat_row = pd.concat([df_a, df_b], ignore_index=True)
print("concat row-wise:\n", concat_row)

# Contoh merge (join by key)
left = pd.DataFrame({"id":[1,2,3], "name":["A","B","C"]})
right = pd.DataFrame({"id":[1,3], "city":["Malang","Jakarta"]})

merged = pd.merge(left, right, on="id", how="left")
print("\nmerge by id (left join):\n", merged)
````
