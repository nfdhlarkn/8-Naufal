## **📌 Soal 2: Ubah Kolom Waktu & Buat Fitur Durasi**

Kolom `pickup` dan `dropoff` masih bertipe string.  
Ubahlah keduanya menjadi datetime, lalu buat kolom baru bernama `trip_duration_min` yang berisi durasi perjalanan dalam **menit**.

Setelah itu:
1. tampilkan 5 baris pertama dari kolom `pickup`, `dropoff`, dan `trip_duration_min`
2. tampilkan rata-rata durasi perjalanan


```python
# Tulis jawabanmu di sini
df['pickup'] = pd.to_datetime(df['pickup'])
df['dropoff'] = pd.to_datetime(df['dropoff'])
df['trip_duration_min'] = (df['dropoff'] - df['pickup']).dt.total_seconds() / 60

# Tampilkan 5 baris pertama
data = df[['pickup', 'dropoff', 'trip_duration_min']]
data.head()
```

```python
# Rata-rata durasi
print("Rata-rata durasi perjalanan: ", data['trip_duration_min'].mean())
```

## **📌 Soal 5: Missing Value**

Kerjakan langkah berikut:
1. Tampilkan jumlah missing value per kolom
2. Buat salinan dataset bernama `df_clean`
3. Pada `df_clean`, isi missing value:
   - kolom `payment` dengan string `"unknown"`
   - kolom `dropoff_borough` dengan string `"Unknown"`
4. Tampilkan lagi jumlah missing value dari dua kolom tersebut setelah diisi

**Catatan:** gunakan `.copy()` supaya dataset asli `df` tetap aman.


```python
# Tulis jawabanmu di sini
# Seluruh jumlah missing value per kolom
df.isna().sum()
```

```python
# Membuat copy dan mengisi missing value menjadi 'unknown'
df_clean = df.copy()
df_clean.loc[df_clean['payment'].isna(), 'payment'] = 'unknown'
df_clean.loc[df_clean['dropoff_borough'].isna(), 'dropoff_borough'] = 'Unknown'
df_clean[['payment', 'dropoff_borough']].isna().sum()
```

## **📌 Soal 6: Data Duplikat**

Pada dataset `taxis`, data duplikat alami jumlahnya sedikit / bahkan bisa tidak ada.  
Supaya bisa latihan, buat dataset baru bernama `df_dupes` dengan cara menambahkan **15 baris acak dari `df_clean`** ke bagian bawah dataset.

Lalu:
1. Hitung jumlah baris duplikat pada `df_dupes`
2. Buang baris duplikat tersebut dan simpan ke `df_nodup`
3. Tampilkan shape sebelum dan sesudah menghapus duplikat


```python
# Tulis jawabanmu di sini
df_dupes = pd.concat([df_clean, df_clean.sample(15, random_state=42)],ignore_index=True)
df_nodup = df_dupes.drop_duplicates()

print("Shape sebelum drop: ", df_dupes.shape)
print("Shape setelah drop: ", df_nodup.shape)
```

## **📌 Soal 7: Feature Engineering Sederhana**

Gunakan `df_nodup`, lalu buat:
1. Kolom `tip_pct` dengan rumus:

$$
\text{tip\_pct} = \frac{\text{tip}}{\text{fare}} \times 100
$$

2. Kolom `is_long_trip`:
   - `"yes"` jika `distance >= 5`
   - `"no"` jika `distance < 5`

3. Kolom `pickup_hour` dari waktu `pickup`

Tampilkan 10 baris pertama dari kolom:
`fare`, `tip`, `tip_pct`, `distance`, `is_long_trip`, `pickup_hour`


```python
# Tulis jawabanmu di sini
df_nodup['tip_pct'] = (df_nodup['tip'] / df_nodup['fare']) * 100
df_nodup['is_long_trip'] = (df_nodup['distance'] >= 5).map({True: 'yes', False: 'no'})
df_nodup['pickup_hour'] = df_nodup['pickup'].dt.hour
df_nodup[['fare', 'tip', 'tip_pct', 'distance', 'is_long_trip', 'pickup_hour']].head(10)
```

## **📌 Soal 8: GroupBy & Aggregation**

Gunakan `df_nodup`, lalu buat ringkasan per `payment` yang berisi:
1. jumlah trip
2. rata-rata `fare`
3. rata-rata `tip`
4. rata-rata `trip_duration_min`

Urutkan hasilnya berdasarkan jumlah trip dari yang terbesar ke yang terkecil.


```python
# Tulis jawabanmu di sini
summary = df_nodup.groupby('payment').agg({'payment': 'count','fare': 'mean','tip': 'mean','trip_duration_min': 'mean'}).rename(columns={'payment': 'trip_count'}).sort_values('trip_count', ascending=False)
summary
```

```python
sum_pick_boro = df_nodup.groupby('pickup_borough').agg({'pickup_borough': 'count'}).rename(columns={'pickup_borough': 'trip_count'}).sort_values('trip_count', ascending=False)
sum_pick_boro
```

## **📌 Soal 9: GroupBy + `transform()`**

Masih menggunakan `df_nodup`, buat kolom baru bernama `fare_vs_payment_mean` dengan rumus:

$$
\text{fare\_vs\_payment\_mean} =
\text{fare} - \overline{\text{fare}}_{\text{(payment yang sama)}}
$$

Artinya, setiap baris dibandingkan terhadap rata-rata `fare` dari metode pembayaran yang sama.

Tampilkan 10 baris pertama dari kolom:
`payment`, `fare`, `fare_vs_payment_mean`


```python
# Tulis jawabanmu di sini
df_nodup['fare_vs_payment_mean'] = df_nodup['fare'] - df_nodup.groupby('payment')['fare'].transform('mean')
df_nodup[['payment', 'fare', 'fare_vs_payment_mean']].head(10)
```

## **📌 Soal 10: Sorting & Menyimpan Hasil**

Buat dataset akhir bernama `final_df` dari `df_nodup` dengan langkah:
1. Urutkan berdasarkan `total` dari terbesar ke terkecil
2. Jika nilai `total` sama, urutkan `trip_duration_min` dari terkecil ke terbesar
3. Simpan hasilnya ke file bernama **`taxis_final.csv`**

Tampilkan 10 baris pertama dari `final_df`.


```python
# Tulis jawabanmu di sini
final_df = df_nodup.sort_values(['total', 'trip_duration_min'], ascending=[False, True])
final_df.head(10)
```

```python
final_df.to_csv('taxis_final.csv', index=False)
```
