## **📌 Soal 4: Boolean Masking & `np.where()`**

Diberikan array suhu berikut:


```python
temps = np.array([29, 31, 35, 28, 33, 27, 30, 36, 32, 26])
print(temps)
```

Kerjakan hal berikut:
1. Ambil semua suhu yang **lebih dari 30**
2. Hitung berapa banyak suhu yang **lebih dari atau sama dengan 32**
3. Buat array label baru bernama `status` dengan aturan:
   - `"panas"` jika suhu > 32
   - `"normal"` jika suhu 29 s.d. 32
   - `"dingin"` jika suhu < 29

**Hint:** untuk poin (3), kamu bisa menggunakan `np.where()` bertingkat.


```python
# Tulis jawabanmu di sini
over30 = temps[temps > 30]
print(over30)
print()

count_over32 = temps[temps >= 32].size
print(count_over32)
print()

status = np.where(temps > 32, "panas", np.where(temps >= 29, "normal", "dingin"))
print(status)
```

## **📌 Soal 5: View vs Copy**

Buat array berikut:


```python
x = np.array([10, 20, 30, 40, 50, 60])
print(x)
```

Lalu lakukan:
1. Buat `view_part` dari elemen indeks 1 sampai 4 menggunakan slicing
2. Buat `copy_part` dari elemen indeks `[1, 2, 3, 4]` menggunakan advanced indexing
3. Ubah elemen pertama `view_part` menjadi `999`
4. Ubah elemen pertama `copy_part` menjadi `-100`
5. Tampilkan `x`, `view_part`, dan `copy_part`

Terakhir, tuliskan dengan singkat:
- mana yang menyebabkan array asli berubah?
- mana yang tidak?


```python
# Tulis jawabanmu di sini
# Tambahkan penjelasan singkat dalam komentar juga.

view_part = x[1:5]
copy_part = x[[1, 2, 3, 4]]
view_part[0] = 999
copy_part[0] = -100

print("Array asli: ", x)
print("Array view_part: ", view_part)
print("Array copy_part: ", copy_part)

# Perubahan pada view_part mengubah isi dari array asli x, hal ini disebabkan array merupakan variabel bertipe data reference sehingga ketika dilakukan slicing variabel view_part akan menyalin referensi dari X yang mana membuat ketika view_part diubah maka x akan ikut berubah
# Perubahan pada copy_part tidak akan mengubah array asli x karena copy_part adalah array yang sepenuhnya baru yang hanya menyalin isi dari variabel x melalui advance indexing
```

## **📌 Soal 6: Statistik Dasar & Missing Value (`NaN`)**

Gunakan array berikut:


```python
values = np.array([12, 15, np.nan, 18, 21, np.nan, 24, 30])
print(values)
```

Tampilkan:
1. Rata-rata biasa dengan `np.mean()`
2. Rata-rata yang mengabaikan `NaN` dengan `np.nanmean()`
3. Standar deviasi yang mengabaikan `NaN` dengan `np.nanstd()`
4. Nilai minimum dan maksimum yang mengabaikan `NaN`

Lalu jelaskan singkat kenapa hasil `np.mean()` bisa berbeda dari `np.nanmean()`.


```python
# Tulis jawabanmu di sini
print("Rata-rata biasa: ", np.mean(values))
print("Rata-rata yang mengabaikan NaN: ", np.nanmean(values))
print("Standar deviasi yang mengabaikan NaN: ", np.nanstd(values))
print("Nilai minimum: ", np.nanmin(values))
print("Nilai maksimum: ", np.nanmax(values))

# Hasil np.mean() berbeda karena np.mean() akan tetap memasukkan dan menjumlahkan nilai NaN (np.nan), seluruh operasi yang melibatkan nilai NaN akan menghasilkan nan
```

## **📌 Soal 7: Broadcasting untuk Normalisasi**

Diberikan matriks fitur berikut:


```python
X = np.array([
    [170, 65, 28],
    [165, 55, 24],
    [180, 75, 30],
    [175, 70, 29]
], dtype=float)

print(X)
```

Anggap setiap kolom merepresentasikan fitur yang berbeda.  
Lakukan normalisasi z-score sederhana dengan rumus:

$$
Z = \frac{X - \text{mean kolom}}{\text{std kolom}}
$$

Tampilkan:
1. Mean per kolom
2. Std per kolom
3. Hasil normalisasi `Z`

**Hint:** gunakan `axis=0` dan manfaatkan broadcasting NumPy.


```python
# Tulis jawabanmu di sini
rata2 = np.mean(X, axis=0)
std = np.std(X, axis=0)
Z = (X - rata2) / std

print("Mean per kolom: ", rata2)
print("Std per kolom: ", std)
print("Hasil normalisasi Z: \n", Z)
```

## **📌 Soal 8: Reshape & Mini Challenge**

Gunakan array 1 dimensi berikut:

Kerjakan hal berikut:

`[[1,2,3,4], [2,3,4,5], ..., [21,22,23,24]]`

**Hint bonus:** boleh gunakan loop **atau** `sliding_window_view` jika kamu sudah tahu.
